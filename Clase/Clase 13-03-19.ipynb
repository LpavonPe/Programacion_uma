```python
def nombres_cargos(x):
    cargo=[]
    nombres=[]
    for i in tabla['Cargo']:
        cargo.append(i)
    for i in range(len(cargo)):
        if cargo[i] == x:
            nombres.append(tabla['Nombre'][i]+" "+tabla['Apellido'][i])
    return nombres
```

```python
def salario_cargo(x):
    if type(x) == int or type(x) == float: 
        equivalencia=162.24
        salario=[]
        nombres=[]
        for i in tabla['Salario']:
            salario.append(i)
        for i in range(len(salario)):
            if salario[i]/equivalencia <= x:
                nombres.append(tabla['Nombre'][i]+" "+tabla['Apellido'][i])
        if len(nombres)>0:        
            return nombres
        else:
            print("Nadie gana menos de $",x)
    else:
        print("La función solo acepta numeros")

```
