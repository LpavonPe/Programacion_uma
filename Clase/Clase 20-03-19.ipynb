```python
tabla=pd.read_excel("Alumnos.xls",sheet_name="Base de Datos estudiantes")
```

```python
nombres_lista=[]
for i in range(len(tabla['NOMBRE'])):
    nombres_lista.append(tabla['NOMBRE'][i])
```

```python
def escuela(especialidad):
    escuelas=[]
    for i in range(len(tabla['ESPECIALIDAD QUE CURSA'])):
        if tabla['ESPECIALIDAD QUE CURSA'][i]==especialidad:
            escuelas.append(tabla['NOMBRE DE LA INSTITUCIÓN EN LA QUE ESTUDIA'][i])
    set_esc=set(escuelas)
    lista_escuelas=list(set_esc)
    df_escuelas=pd.DataFrame(data=lista_escuelas)
    with pd.ExcelWriter("Escuelas.xls") as archivo:
        df_escuelas.to_excel(archivo)
    
    
```

```python
def nombres_esp(especialidad):
    nombres=[]
    for i in range(len(tabla['ESPECIALIDAD QUE CURSA'])):
        if tabla['ESPECIALIDAD QUE CURSA'][i]==especialidad:
            nombres.append(tabla['NOMBRE'][i])
    df_nom=pd.DataFrame(data=nombres)
    with pd.ExcelWriter("Nombres.xls") as archivo:
        df_nom.to_excel(archivo)
    
    
```
